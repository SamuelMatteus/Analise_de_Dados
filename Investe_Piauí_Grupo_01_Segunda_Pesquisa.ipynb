```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```


Para essa nova fase, utilizaremos os dados apurados com base nas preferências de consumo dos clientes, utilizando também elementos demográficos e comportamentais deles, o objetivo é construir um perfil com base nos dados para criar uma abordagem de venda mais efetiva.
---



```python
dataframe = pd.read_csv('Segunda-Pesquisa-Dados-Tratados.csv')
print(dataframe.head())
```

Para essa segunda pesquisa, o arquivo csv foi tratado previamente via Excel e Power Query

```python
dataframe.describe()
```

Para essa análise, não se faz necessário o número dos clientes, portanto desconsideramos o uso da mesma, mas sem excluir os dados.

```python
colunas_dataframe = dataframe.columns
print(colunas_dataframe)
```

```python
contagem_por_sexo = dataframe['Sexo'].value_counts()
plt.pie(contagem_por_sexo, labels=contagem_por_sexo.index, autopct='%1.1f%%', startangle=90)
plt.title('Frequência de Consumo por Sexo')
plt.show()
```

Nosso público agora na pesquisa se confirma como maioria feminina.

```python
media_idade_por_genero = dataframe.groupby('Sexo')['Informe sua idade (Apenas Números)'].mean()

plt.bar(media_idade_por_genero.index, media_idade_por_genero, color=['pink', 'blue'])

plt.title('Média de Idade por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Anos')

plt.show()
```

Com a média de idade estabelecida podemos considerar nosso público, pessoas do sexo masculino e feminino na média dos 25/26 anos.

```python
freq_passagem = dataframe['Com que frequência você consome Açaí?'].value_counts()

freq_passagem.plot(kind='bar', color=['green', 'orange', 'green', 'red', 'green'])
plt.title('Frequência de Consumo de Açaí')
plt.xlabel('Período')
plt.ylabel('Contagem de Pessoas')
plt.show()
```

Deduzimos também o maior fluxo de consumo de açaí é mensal. Agora vamos ver quanto desse nosso publico mora ou frequenta a zona leste.

```python
moradores_zona_leste = dataframe['Você mora por essa região?'].value_counts()
plt.pie(moradores_zona_leste, labels=moradores_zona_leste.index, autopct='%1.1f%%', startangle=90)
plt.title('Moradores da Zona Leste')
plt.show()
```

Constatado que a maior parte do público da pesquisa não mora na zona leste, mas porque esses dados ainda são relevantes?

```python
frequentadores_zona_leste = dataframe['Com que frequência você passa pela área da Avenida João XXIII, no balão do São Cristóvão?'].value_counts()
plt.pie(frequentadores_zona_leste, labels=frequentadores_zona_leste.index, autopct='%1.1f%%', startangle=90)
plt.title('Pessoas que passam pela região')
plt.show()
```

Levando em conta que temos 85% de pessoas que passam pela região, eles também se tornam relevantes para a pesquisa.

```python
horario_frequencia = dataframe['Em qual horário você costuma passar ou ficar pela região?'].value_counts()
plt.pie(horario_frequencia, labels=horario_frequencia.index, autopct='%1.1f%%', startangle=90)
plt.title('Horário em que as pessoas passam pela região')
plt.show()
```

Conforme o gráfico de horários, temos uma grande fatia que passa no horário comercial, e outra que passa pela noite, pós horário comercial.

E aproveitando esses dados, vamos agora ver também em qual horário o público da pesquisa prefere comer o açaí:

```python
preferencia_horario_consumo = dataframe['E em qual horário? Escolha o intervalo que mais combina!'].value_counts()
plt.pie(preferencia_horario_consumo, labels=preferencia_horario_consumo.index, autopct='%1.1f%%', startangle=90)
plt.title('Frequência de Consumo por Sexo')
plt.show()
```

Conforme o gráfico, temos um público que prefere um consumo maior na noite, logo seguido de um público que prefere tarde também. Sendo assim nosso público alvo são homens e mulheres, na média de 25/26 anos, que mesmo não morando na zona leste, passam por ela pelo menos uma vez na semana, onde passam em sua maioria nos horários comercial e pós comercial

Agora vamos ver se esse mesmo público gosta e valida o Açaí no Grau:

```python
aceitacao = dataframe['Se você fosse dar uma nota para o Açaí no Grau, qual seria?'].value_counts()
plt.pie(aceitacao, labels=aceitacao.index, autopct='%1.1f%%', startangle=90)
plt.title('Notas do público - Açaí no Grau')
plt.show()
```

```python
aceitacao_gelato = dataframe['De 0 a 10, qual a chance de você comprar algum produto da nossa linha de picolés e gelatos?'].value_counts()
plt.pie(aceitacao_gelato, labels=aceitacao_gelato.index, autopct='%1.1f%%', startangle=90)
plt.title('Aceitação - Gelatos - Açaí no Grau')
plt.show()
```

Tanto a linha de açaí como gelatos tem aceitação pela maior parte dos usuários que responderam a pesquisa.

```python
colunas_dataframe = dataframe.columns
print(colunas_dataframe)
```

```python
mais_preferidos = dataframe['O que mais gosta do Açaí no Grau?'].value_counts()
plt.pie(mais_preferidos, labels=mais_preferidos.index, autopct='%1.1f%%', startangle=90)
plt.title('O que mais preferem no Açaí no Grau?')
plt.show()
```

Quando colocamos as preferências dos produtos, o Sabor juntamente com a Localização lideram no gráfico, logo em seguida, apenas do sabor e complementos.Aproveitamos e também sugerimos uma pergunta com base no desconto previsto pelo chefe de operações da rede, e esse é o resultado:

```python
sugestao_promocao = dataframe['Se você ganhasse 10% de desconto no seu pedido levando um amigo comprar um açaí, qual seria sua chance de comprar com a gente na nossa nova unidade?'].value_counts()
plt.pie(sugestao_promocao, labels=sugestao_promocao.index, autopct='%1.1f%%', startangle=90)
plt.title('Indicação em caso de Promoção de 10%')
plt.show()
```

Por mais que tenhamos um resultado bem dinâmico, as notas 10 e 9 ainda representam 75% de todos os clientes.
