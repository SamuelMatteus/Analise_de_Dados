```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Resultados.csv')
print(df.head())
```

```python
df.describe()
```

```python
colunas = df.columns
print(colunas)
```

Contagem : Sexo

```python
contagem_sexo = df['Qual seu sexo?'].value_counts()
plt.pie(contagem_sexo, labels=contagem_sexo.index, autopct='%1.1f%%', startangle=90)
plt.title('Frequência de Consumo de Açaí')
plt.show()
```

Média de idade dos consumidores

```python
media_idade_por_genero = df.groupby('Qual seu sexo?')['Quantos anos você tem?'].mean()

plt.bar(media_idade_por_genero.index, media_idade_por_genero, color=['pink', 'blue'])

plt.title('Média de Idade por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Anos')

plt.show()
```

Períodos da semana em que consomem:

```python
consumo = df['Em que período da semana você costuma comer açaí?'].value_counts()
print(consumo)

consumo.plot(kind='bar', color=['brown', 'blue'])
plt.title('Pessoas X Período de Consumo')
plt.xlabel('Período')
plt.ylabel('Contagem de Pessoas')
plt.show()
```

Frequência de consumo:

```python
freq_consumo = df['Quantas vezes no mês você consome açaí'].value_counts()
plt.pie(freq_consumo, labels=freq_consumo.index, autopct='%1.1f%%', startangle=90)
plt.title('Frequência de Consumo de Açaí')
plt.show()
```

```python
hora_consumo = df['Qual horário do dia você costuma pedir açaí?'].value_counts()
plt.pie(hora_consumo, labels=hora_consumo.index, autopct='%1.1f%%', startangle=90)
plt.title('Horários de Consumo de Açaí')
plt.show()
```

```python
consumo_tamanho = df['Qual tamanho de açaí você costuma pedir?'].value_counts()
plt.pie(consumo_tamanho, labels=consumo_tamanho.index, autopct='%1.1f%%', startangle=90)
plt.title('Consumo de Açaí por Tamanho')
plt.show()
```

Com base nos dados temos uma conclusão que nosso público tem uma média de idade na casa dos 26 anos, consumindo o produto em torno de uma a duas vezes no mês em sua maioria, nos fins de semana e a noite onde a preferência maior é pelo tamanho médio do produto.
